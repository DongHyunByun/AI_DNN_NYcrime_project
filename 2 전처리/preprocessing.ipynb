```python
# 요일반환 : (월~일)=>(0~6)
# 월변환 : (1~12)
import datetime
def getDay(day):
    m=int(day[:2])
    d=int(day[3:5])
    y=int(day[6:])
    r=datetime.date(y,m,d).weekday()
    
    return [m-1,d-1,r]



#시간 범주화 (0~24)
def getTime(time):
    if time[1]==":":
        h=int(time[0])
    else:
        h=int(time[:2])
    
    return h

    
# 범죄유형 범주화 (0~6)
def getCrimeCode(ky_cd):
    ky_cd=int(ky_cd)
    # 교통법 관련
    if ky_cd in [347,353,102,348]:
        return 0
    # 절도
    elif ky_cd in [232,342,109,343,111,107,112,341,110,105,231]:
        return 1
    # 사기,위조,도박
    elif ky_cd in [358,238,113,101,340,571,350,122]:
        return 2
    # 약물, 술
    elif ky_cd in [346,235,117]:
        return 3
    # 폭행, 납치
    elif ky_cd in [236,118,578,106,352,344,363,355,124]:
        return 4
    # 성범죄
    elif ky_cd in [234,115,356,104,116,233]:
        return 5
    # 기타
    elif ky_cd in [121,351,677,126,678,364,103,359,455,354,572,125]:
        return 6
    else:
        return 7

def getStreet(a):
    try:
        a=int(a)
        if 1<=a<10:
            return 0
        elif 10<=a<20:
            return 1
        elif 20<=a<30:
            return 2
        elif 30<=a<40:
            return 3
        elif 40<=a<50:
            return 4
        elif 50<=a<60:
            return 5
        elif 60<=a<70:
            return 6
        elif 70<=a<80:
            return 7
        elif 80<=a<90:
            return 8
        elif 90<=a<100:
            return 9
        elif 100<=a<110:
            return 10
        else:
            return 11  
    except:
        return 12
    
```

```python
import os
import pandas as pd
import numpy as np

folder="each_city"
subdir_names=os.listdir(folder)
for file_name in subdir_names:
    df=pd.DataFrame(columns=["month","day","dow","time","crime_code","street"])
    csv=pd.read_csv(folder+"\\"+file_name) 
    for i in range(len(csv)):
        crimeNum=getCrimeCode(csv["KY_CD"][i])
        stNum=getStreet(csv["ADDR_PCT_CD"][i])
        if crimeNum!=7 and stNum!=12:
            m,d,r=getDay(csv["CMPLNT_FR_DT"][i])
            t=getTime(csv["CMPLNT_FR_TM"][i])
            c=crimeNum
            st=stNum
            df=df.append({"month":m,"day":d,"dow":r,"time":t,"crime_code":c,"street":st},ignore_index=True)
    df.to_csv("preprossed\\"+file_name[:-4]+"W.csv",mode='w')
```
